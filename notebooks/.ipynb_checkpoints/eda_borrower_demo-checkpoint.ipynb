```python
df
```

```python
# df.loan_status.unique()

default_statuses = ['Charged Off', 'Default', 'Does not meet the credit policy. Status:Charged Off']

# Create numeric target
df['loan_status'] = df['loan_status'].apply(lambda x: 1 if x in default_statuses else 0)
```

```python
# 2. Instantiate the EDA class:

eda = BorrowerProfileEDA(df, target_col="loan_status")
```

```python
# 3. Run the pipeline and inspect the results:

report = run_borrower_eda_pipeline(eda)
```
